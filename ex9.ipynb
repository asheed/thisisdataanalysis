```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import sys
```

```python
with open('bs연습문제.html', encoding='UTF-8') as ex:
    soup = BeautifulSoup(ex, 'html.parser')

# print(soup.prettify())
    
img_src = soup.find('div', 'flex_grid credits search_results').select('img')

img_src_2017 = []
img_src_2018 = []
# print( img_src )

for i in img_src:
    # print(i['src'])
    if i['src'].split('/')[4] == '2017':
        img_src_2017.append(i['src'])
    elif i['src'].split('/')[4] == '2018':
        img_src_2018.append(i['src'])

print("2017년 데이터는 총 %s건 입니다." % len(img_src_2017))
for i in img_src_2017:
    print(i)

print("2018년 데이터는 총 %s건 입니다." % len(img_src_2018))
for i in img_src_2018:
    print(i)
```
